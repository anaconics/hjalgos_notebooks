```python
# {vbt.ADVSTEX,
# vbt.ATR,
# vbt.BBANDS,
# vbt.HSTD,
# vbt.MACD,
# vbt.IADVSTEX,
# vbt.IRPROBEX,
# vbt.ISTEX,
# vbt.MA,
# vbt.OBV,
# vbt.RAND,
# vbt.RPROB,
# vbt.RPROBEX,
# vbt.RPROBEX RSI,
# vbt.RSI,
# vbt.STEX,
# vbt.STOCH}

def calc_port(mydf1):
    close = mydf1['close']
    high = mydf1['high']
    low = mydf1['low']
    
    fast_ma, slow_ma = vbt.MA.run_combs(
        close, window=np.arange(2, 101), r=2, 
        short_names=['fast', 'slow'])
    entries_ma = fast_ma.ma_above(slow_ma, crossed=True)
    exits_ma = fast_ma.ma_below(slow_ma, crossed=True)
    portfolio_ma = vbt.Portfolio.from_signals(close, entries_ma, exits_ma, fees=0.001)
    
    fast_stoch, slow_stoch = vbt.STOCH.run_combs(high,low,close, 
                                           k_window=np.arange(2, 101), 
                                           d_window=np.arange(2, 101), 
                                           r=2,
                                           short_names=['fast', 'slow'])
    
    entries_stoch = fast_stoch.percent_d_above(fast_stoch.percent_k, crossed=True)
    exits_stoch = fast_stoch.percent_d_below(fast_stoch.percent_k, crossed=True)
    
#     entries_stoch = slow_stoch.percent_d_above(slow_stoch.percent_k, crossed=True)
#     exits_stoch = slow_stoch.percent_d_below(slow_stoch.percent_k, crossed=True)
    
    
    portfolio_stoch = vbt.Portfolio.from_signals(close, entries_stoch, exits_stoch, fees=0.001)
    
    return [portfolio_ma,portfolio_stoch]
    # Get total return, reshape to symmetric matrix, and plot the whole thing
    # fig = portfolio.total_return().vbt.heatmap(
    #     x_level='fast_window', y_level='slow_window', symmetric=True,
    #     trace_kwargs=dict(colorbar=dict(title='Total return', tickformat='%')))
    # fig.show()
```
